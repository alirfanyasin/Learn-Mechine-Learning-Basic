# Seaborn

Seaborn adalah library visualisasi data yang luar biasa untuk grafik statistik yang diplot dengan Python. Ini memberikan gaya default yang indah dan palet warna untuk membuat plot statistik lebih menarik. Itu dibangun di atas perpustakaan matplotlib dan juga terintegrasi erat dengan struktur data dari panda. 

Topik bahasan:
* Seaborn Color Palette
* Fungsi replot
* Bar Plot
* Count Plot
* Point Plot
* Joint Plot
* KDE Plot
* Heatmap
* Pair Plot

## Bar Plot
### Vertical barplot

````python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
 
# baca dataset
titanic = sns.load_dataset('titanic')
titanic.head()
````

````python
# plot
sns.barplot(x = 'embark_town', y = 'age', data = titanic, palette = 'Accent',ci=None)
plt.show()
````

````python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns


# load dataset
titanic = sns.load_dataset('titanic')
# create plot
sns.barplot(x = 'sex', y = 'survived', hue = 'class', data = titanic,
            palette = 'RdPu_r',
            order = ['male', 'female'],  
            capsize = 0.05,             
            saturation = 8,             
            errcolor = 'gray', errwidth = 2,  
            ci = 'sd'  
            )
plt.legend()
plt.show()
````

### Horizontal barplot

````python
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
  
sns.barplot(x = 'age', y = 'embark_town', data = titanic,
            palette = 'PuRd', orient = 'h')
plt.show()
````

## Joint Plot

````python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

sns.set_style("dark")
tips=sns.load_dataset('tips')
sns.jointplot(x='total_bill', y='tip',data=tips)
````

````python
# Tampilkan perkiraan kepadatan kernel alih-alih plot sebar dan histogram
sns.jointplot(x='total_bill', y='tip', data=tips, kind='kde', fill=True)
````

````python
import seaborn as sns
import matplotlib.pyplot as plt
%matplotlib inline

iris=sns.load_dataset('iris')
sns.jointplot(x='sepal_length', y='petal_length',data=iris)
````

````python
sns.jointplot(x='sepal_length', y='petal_length',data=iris, kind='kde', fill=True)
````

````python
sns.jointplot(data=iris, x='sepal_length', y='petal_length', hue="species")
````

## KDE (Kernel Density Estimate) plot

````python
iris = sns.load_dataset("iris")
iris.head()
````

````python
# plot
import matplotlib.pyplot as plt
%matplotlib inline

sns.kdeplot(iris.loc[(iris['species']=='setosa'), 
            'sepal_length'], color='b', shade=True) 
sns.kdeplot(iris.loc[(iris['species']=='virginica'), 
            'sepal_length'], color='r', shade=True)
sns.kdeplot(iris.loc[(iris['species']=='versicolor'), 
            'sepal_length'], color='g', shade=True)

plt.legend(iris['species'].unique().tolist())
````

````python
sns.kdeplot(
   data=iris, x="sepal_length", hue="species",
   fill=True, common_norm=False, palette="crest",
   alpha=.5, linewidth=0,
)
````

````python
sns.kdeplot(
    data=iris, x="sepal_length", y="petal_width", hue="species", fill=True,
)
````

## Pair Plot

````python
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline

df = sns.load_dataset('iris')
sns.set_style("ticks")
sns.pairplot(df,hue = 'species',diag_kind = "kde",kind = "scatter",palette = "husl")
plt.show()
````

````python
sns.pairplot(df, hue="species", diag_kind="hist")
````

## Ukuran Gambar dan Subplot

````python
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline

df = sns.load_dataset('iris')
df.head()
````

````python
sns.set_theme(color_codes=True)
sns.histplot(x='sepal_length', data=df, hue='species', kde = False)
````

````python
# ubah ukuran
fig, ax = plt.subplots(figsize=(10,6))
sns.histplot(x='sepal_length', data=df, hue='species', kde = False, ax=ax)
````

````python
# subplot
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2,figsize=(12,9))

sns.histplot(x='sepal_length', data=df, hue='species', kde = False, ax = ax1)
ax1.set_title('sepal_length')
sns.histplot(x='sepal_width', data=df, hue='species', kde = False, ax=ax2)
ax2.set_title('sepal_width')
sns.histplot(x='petal_length', data=df, hue='species', kde = False, ax=ax3)
ax3.set_title('petal_length')
sns.histplot(x='petal_width', data=df, hue='species', kde = False, ax=ax4)
ax4.set_title('petal_width')

fig.tight_layout()
````
