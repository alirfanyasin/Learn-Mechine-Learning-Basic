# Boolean Indexing

```python
import pandas as pd
```

```python
pd.__version__
```

```python
df = pd.read_csv('dataset/iris.csv')
df
```

```python
df.head()
```

```python
# Filter dataframe berdasarkan kondisi nilai pada satu kolom
df[df['Sepal.Length'] > 7.5]
```

```python
# Filter dataframe berdasarkan beberapa kondisi nilai pada satu 

df[df['Species'].isin(['versicolor', 'virginica'])]
```

```python
# Filter berdasarkan beberapa kondisi pada beberapa kolom menggunakan operator AND
df[(df['Sepal.Length'] > 7.5) & (df['Sepal.Width'] > 3) ] # AND operator
```

```python
# Filter berdasarkan beberapa kondisi pada beberapa kolom menggunakan operator OR
df[(df['Sepal.Length'] > 7.5) | (df['Sepal.Width'] > 3)] # OR operator
```
