```python
df = pd.read_csv('dataset/iris.csv')
df
```

```python
# Mengganti spesifik nama kolom 
df.rename(columns={'Sepal.Length'  : 'Sepal_Length'}, inplace=True)
df
```

```python
# Mengganti semua nama kolom
df.columns = ['sepal_length', 'sepal_width', 'petal_length', 'petal_width', 'species']
df
```

```python
# Data Dummy

raw_data = {
    'first_name' : ['Amy', 'Amy', 'Jason', 'Nick', 'Stephen', 'Amy'],
    'last_name' : ['Jackson', 'J', 'Miller', 'Milner', 'L', 'J'],
    'age' : [42,42,36,24,24,42]
           }

df = pd.DataFrame(raw_data, columns=['first_name', 'last_name', 'age'])
df
```

```python
# Hapus duplikasi
df.duplicated()
```

```python
df.drop_duplicates()
```

```python
df.drop_duplicates(['first_name'], keep='first')
```

```python
# Membuat kolom baru dari kolom yang ada 
df['age_plus_5'] = df['age'] + 5
df
```

```python
# Membuat kolom baru dari dua element kolom 
df['full_name'] = df['first_name'] + '_' + df['last_name']
df
```

```python
# Menambahkan list untuk kolom baru pada dataframe
df['gender'] = pd.Series(['F','F','M','M','M','F'])
df
```
