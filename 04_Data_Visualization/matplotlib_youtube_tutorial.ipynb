# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8]
y = [1,3,6,3,7,8,4,2]

fig, ax = plt.subplots() # membuat sebuah figur dan 
ax.plot(x,y) # melakukan plotting data pada axes

ax.set_ylabel('Label Y')
ax.set_xlabel('Label X')
fig.suptitle('Sample Figure Title')
```

## Object Oriented Style untuk melakukan plotting

```python
x = np.linspace(0, 2, 100)
x
```

```python
fig, ax = plt.subplots()

# plotting 3 varian data
ax.plot(x,x, label='linear')
ax.plot(x,x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')

ax.set_xlabel('x label', fontsize=16)
ax.set_ylabel('y label', fontsize=16)
ax.set_title('Simple Plot', fontsize=20)


ax.legend()
```

```python
plt.plot(x,x, label='linear')
plt.plot(x,x**2, label='quadratic')
plt.plot(x,x**3, label='cubic', linewidth=6)

plt.xlabel('x label', fontsize=16)
plt.ylabel('y label', fontsize=16)
plt.title('Sampel Title', fontsize=20)

plt.legend()
```

## Plotting dengan keyword

```python
data = {
  'a' : np.arange(50), # 0 - 49
  'c' : np.random.randint(0, 50, 50), # generate bilangan random sebanyak 50 bilangan, mulai dari angka 0 - 50
  'd' : np.random.randn(50) # generate bilangan random sebanyak 50 bilangan
}

data['b'] = data['a'] + 10 * np.random.randn(50)
data['d'] = np.abs(data['d']) * 10

data
```

```python
plt.scatter('a', 'b', data=data)
plt.xlabel('entry a')
plt.ylabel('entry y')
plt.show()
```

```python
plt.scatter('a', 'b', c='c', s='d', data=data)
plt.xlabel('entry a')
plt.ylabel('entry b')
plt.show()

# catatan:
# c = adalah color
# s = adalah size
```

## Multiple Subplots

```python
x = np.arange(0.0, 2.0, 0.01)
x
```

```python
s = np.sin(2 * np.pi * x)
s
```

```python
fig, ax = plt.subplots()

ax.plot(x, s)
ax.set(
  xlabel='Nilai x',
  ylabel='Nilai sin(x)',
  title='Visualisasi nilai sin'
)
ax.yaxis.grid()
plt.show()
```

## Multiple Subplots Object Oriented Style

```python
x1 = np.linspace(0.0, 5.0, 100)
x2 = np.linspace(0.0, 2.0, 100)

y1 = np.cos(2 * np.pi * x1)
y2 = np.cos(2 * np.pi * x2)
```

```python
fig, (ax1, ax2) = plt.subplots(2, 1)
fig.suptitle('Multiple Subplots')

ax1.plot(x1, y1, 'ro-')
ax1.set_ylabel('Nilai cos(x1)')
ax1.yaxis.grid()


ax2.plot(x2, y2, 'g.-')
ax2.set_ylabel('Nilai cos(x2)')

ax2.set_xlabel('Nilai X')

plt.show()
```

## Multiple Subplots dengan pyplot Style

```python
plt.subplot(2,1,1) # parameter ketiga merupakan sebuah inisialisasi dari plot 1
plt.plot(x1, y1, 'yo-')
plt.ylabel('Nilai $cos(x1)$')
plt.title('Multiple Subplot')

plt.subplot(2,1,2)
plt.plot(x2,y2, 'b*')
plt.ylabel('Nilai $cos(x2)$')
plt.xlabel('Nilai $X$')

plt.show()
```

## Histogram

```python
mu, sigma = 100, 15 # nilai mean dan nilai standard deviation

x = mu + sigma * np.random.randn(10000) # normal distribution
x
```

```python
x.shape
```

```python
plt.hist(x, bins=50, color='green', alpha=0.75) # hist(sample data, bins)

plt.xlabel('Sumbu X')
plt.ylabel('Sumbu Y')
plt.title('Contoh Histogram')

plt.text(45, 500, r'$\mu=100,\ \sigma=15$') # menampilkan rumusnya
plt.grid()
plt.show()
```

## Histogram dengan OO Style

```python
fig, ax = plt.subplots()

ax.hist(x, bins=50, alpha=0.70) 
ax.set(ylabel='Sumbu Y', xlabel='Sumbu X')
ax.set_title('Contoh Histogram')

ax.text(45,500, r'$\mu=100,\ \sigma=15$')
ax.grid()
plt.show()
```

## Simple Bar Plot

```python
data = [25, 45, 50 ,125, 225]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data, 0.5) # bar(x, y)
plt.grid()

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot', fontsize=18)
plt.show()
```

## Pengaturan grid dan color pada bar plot

```python
plt.bar(kategori, data, color='g', alpha=0.75) # bar(x, y)
plt.grid(axis='y', linestyle='--', linewidth=2, alpha=0.50)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Bar Plot', fontsize=18)
plt.show()
```

## Horizontal Bar Plot

```python
# plt.barh(kategori, data, color='g', alpha=0.75) # bar(x, y)
plt.barh(kategori[::-1], data[::-1], color='g', alpha=0.75) # bar(x, y)
plt.grid(axis='x', linestyle='--', linewidth=2, alpha=0.50)

plt.ylabel('Kategori')
plt.xlabel('Jumlah')
plt.title('Contoh Horizontal Bar Plot', fontsize=18)
plt.show()
```

## Stacked Bar

```python
data1 = [25, 85, 75, 40, 60]
data2 = [40, 35, 20, 55, 10]
kategori = ['A', 'B', 'C', 'D', 'E']
```

```python
plt.bar(kategori, data1, label='Data 1', color='r')
plt.bar(kategori, data2, label='Data 2', bottom=data1)

plt.legend(loc='best')

plt.grid(linestyle='--', axis='y', linewidth=1, alpha=0.50)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Contoh Stacked Bar')
plt.show()
```

## Groupped Bar Plot

```python
data1 = [25,85,75,40,60]
data2 = [40,35,20,55,10]
kategori = ['A','B','C','D','E']
x = np.arange(len(kategori))
x
```

```python
width = 0.25 # mulai dari 0.0 - 0.5

plt.bar(x-width/2, data1, width, label='Data 1')
plt.bar(x+width/2, data2, width, label='Data 2')

plt.xticks(x, kategori) # mengubah dari angka ke kategori atau dari x ke kategori

plt.grid(
  linestyle='--',
  axis='y',
  alpha=0.40,
  linewidth=1
)

plt.xlabel('Kategori')
plt.ylabel('Jumlah')
plt.title('Groupped Bar Plot')

plt.legend()

plt.show()
```

## Pie Plot

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215, 130, 245, 210, 225]
```

```python
plt.pie(data, 
        labels=kategori, # optional
        autopct='%.1f%%', # optional dan untuk menampilkan 
        startangle=90)
plt.title('Pie Chart')
plt.show()
```

## Pie Plot - Pengaturan Legend dan Color

```python
kategori = ['A', 'B' ,'C', 'D', 'E']
data = [215,130,245,210,225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
```

```python
plt.pie(data, 
        labels=kategori,
        colors=warna, # mengambil data warna yang sudah di set
        autopct='%1.1f%%',
        startangle=90)

plt.legend()
plt.title('Pie Chart')

plt.show()
```

## Exploded Pie Plot

```python
kategori = ['A', 'B', 'C', 'D', 'E']
data = [215,130,245,210,225]
warna = ['pink', 'cyan', '#e03364', 'yellowgreen', 'skyblue']
explode_var = [0., 0., 0.2, 0., 0.08]
```

```python
plt.pie(data,
        labels=kategori,
        colors=warna,
        autopct='%1.1f%%',
        startangle=90,
        explode=explode_var)

plt.title('Exploded Pie Chart')
plt.show()
```

## Table Plot / Table Chart

```python
data_uas = [
  ['Bejo', 70],
  ['Tejo', 83],
  ['Cecep', 62],
  ['Wati', 74],
  ['Kartil', 71],
]

data_uas
```

```python
fig, ax = plt.subplots()

table = plt.table(cellText=data_uas, loc='center')
table.set_fontsize(14)
table.scale(1,4) # ukuran kolom dan baris

ax.axis(False) # menghilangkan axis

plt.show()
```

## Table Plot dengan Pyplot Style

```python
table = plt.table(cellText=data_uas, loc='center')
table.set_fontsize(14)
table.scale(1,4)

# untuk menghilangkan axis
ax = plt.gca()
ax.axis(False)

plt.show()
```
