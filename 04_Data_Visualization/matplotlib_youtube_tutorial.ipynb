# Matplotlib

```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

```python
x = [1,2,3,4,5,6,7,8]
y = [1,3,6,3,7,8,4,2]

fig, ax = plt.subplots() # membuat sebuah figur dan 
ax.plot(x,y) # melakukan plotting data pada axes

ax.set_ylabel('Label Y')
ax.set_xlabel('Label X')
fig.suptitle('Sample Figure Title')
```

## Beberapa cara untuk membuat `figure` dan `axes`

```python
# figure tanpa axes
fig = plt.figure()
```

```python
# figure dengan axes
fig, ax = plt.subplots()
```

```python
# figure dengan 2x3 grid axes
fig, ax = plt.subplots(2, 3) 
```

## Object Oriented Style untuk melakukan plotting

```python
x = np.linspace(0, 2, 100)
x
```

```python
fig, ax = plt.subplots()

# plotting 3 varian data
ax.plot(x,x, label='linear')
ax.plot(x,x**2, label='quadratic')
ax.plot(x, x**3, label='cubic')

ax.set_xlabel('x label', fontsize=16)
ax.set_ylabel('y label', fontsize=16)
ax.set_title('Simple Plot', fontsize=20)


ax.legend()
```

```python
plt.plot(x,x, label='linear')
plt.plot(x,x**2, label='quadratic')
plt.plot(x,x**3, label='cubic')

plt.xlabel('x label', fontsize=16)
plt.ylabel('y label', fontsize=16)
plt.title('Sampel Title', fontsize=20)

plt.legend()
```
