# Seaborn

Seaborn adalah library visualisasi data yang luar biasa untuk grafik statistik yang diplot dengan Python. Ini memberikan gaya default yang indah dan palet warna untuk membuat plot statistik lebih menarik. Itu dibangun di atas perpustakaan matplotlib dan juga terintegrasi erat dengan struktur data dari panda. 

Topik bahasan:
* Seaborn Color Palette
* Fungsi replot
* Bar Plot
* Count Plot
* Point Plot
* Joint Plot
* KDE Plot
* Heatmap
* Pair Plot

## KDE Plot

````python
iris = sns.load_dataset('iris')
iris.head()
````

````python
import matplotlib.pyplot as plt
%matplotlib inline

sns.kdeplot(iris.loc[(iris['species'] == 'setosa'), 'sepal_length'], color='b', fill=True)
sns.kdeplot(iris.loc[(iris['species'] == 'virginica'), 'sepal_length'], color='r', fill=True)
sns.kdeplot(iris.loc[(iris['species'] == 'versicolor'), 'sepal_length'], color='g', fill=True)

plt.legend(iris['species'].unique().tolist())
plt.show()
````

````python
sns.kdeplot(data=iris, x='sepal_length', hue='species',
            fill=True, common_norm=False, palette='crest', alpha=0.5, linewidth=0)

plt.show()
````

````python
sns.kdeplot(data=iris, x='sepal_length', y='petal_width', hue='species', fill=True)
plt.show()
````

## Heatmap

````python
flights = sns.load_dataset('flights')
flights.head()
````

````python
# jadikan pivot 
flights =  flights.pivot(index='month', columns='year', values='passengers')
flights
````

````python
sns.heatmap(flights, linewidths=.5, cmap='Reds')
plt.show()
````

## Ukuran Gambar dan Subplot

````python
import seaborn as sns
from matplotlib import pyplot as plt
%matplotlib inline

df = sns.load_dataset('iris')
df.head()
````

````python
sns.set_theme(color_codes=True)
sns.histplot(x='sepal_length', data=df, hue='species', kde = False)
plt.show()
````

````python
# ubah ukuran
fig, ax = plt.subplots(figsize=(10,6))
sns.histplot(x='sepal_length', data=df, hue='species', kde = False, ax=ax)
plt.show()
````

````python
# subplot
fig, ((ax1, ax2), (ax3, ax4)) = plt.subplots(2,2,figsize=(12,9))

sns.histplot(x='sepal_length', data=df, hue='species', kde = False, ax = ax1)
ax1.set_title('sepal_length')
sns.histplot(x='sepal_width', data=df, hue='species', kde = False, ax=ax2)
ax2.set_title('sepal_width')
sns.histplot(x='petal_length', data=df, hue='species', kde = False, ax=ax3)
ax3.set_title('petal_length')
sns.histplot(x='petal_width', data=df, hue='species', kde = False, ax=ax4)
ax4.set_title('petal_width')

fig.tight_layout()
plt.show()
````
